## matplotlib

This library works very well for simpler charts, via the `pyplot` module, which maintains an internal state on which we build a viz piece by piece, and can save via `savefig()` and display with `show()`, as well as including the magic function `%matplotlib inline` in IPython/Jupyter to display plots in the notebook.

```python
%matplotlib inline

from matplotlib import pyplot as plt

# lists with data
years = [1950, 1960, 1970, 1980, 1990, 2000, 2010]
gdp = [300.2, 543.3, 1075.9, 2862.5, 5979.6, 10289.7, 14958.3]

# green line chart of GDP vs. Year
plt.plot(years, gdp, color = "green",
         marker = "o", linestyle = "solid")
# add title
plt.title("Nominal GDP")
# add label to y-axis
plt.ylabel("Billions of $")
plt.show()
```

To make more "publication-friendly" plots, it would require some more complicated logic and coding, with more customizations, which can be seen in `matplotlib`'s documentation.

For now, we can make:
### Bar charts

These are good for showing how quantities vary for different *discrete* categories (quantitative and qualitative data)

```python
movies = ["Annie Hall", "Ben-Hur", "Casablanca",
          "Gandhi", "West Side Story"]
num_oscars = [5, 11, 3, 8, 10]

# add .1 to the left coordinate of each data point
# bars are .8 width by default
# add .1 to each index in enumeration of movies
xs = [i + .1 for i, _ in enumerate(movies)]
print(list(xs))
```

```python
# plt.bar(x-coordinates)
plt.bar(xs, num_oscars)
plt.title("My Favorite Movies")
plt.ylabel("Number of Academy Awards")

# label ticks on x-axis with movie names and set at bar centers
plt.xticks([i+.5 for i, _ in enumerate(movies)], movies)
plt.show()
```

### Histograms

These are bar charts of bucketed numeric values to show distributions.

```python
from collections import Counter

grades = [83,95,91,87,70,0,85,82,100,67,73,77,0]

# function = for a given grade, perform integer division 
# then multiply by 10 to get buckets of size = 10
decile = lambda grade: grade // 10 * 10

# create range for histogram bins
histogram = Counter(decile(grade) for grade in grades)

for k,v in histogram.items():
    print(k,v)
```

```python
plt.bar(histogram.keys(), 
       # get height for bars via Counter `histogram`'s values
        histogram.values(),
        # make width of bars = 8
        # will leave space between bars since buckets = size 10
       8)

# make x-axis in range (-5,105) and y-axis range = 0-5
# so it fully shows all bars
# bad form to not start y @ 0 = misleading
plt.axis([-5,105,0,5])

# make x-ticks every 10
plt.xticks([10*i for i in range(11)])
plt.title("Distribution of Exam 1 Grades")
plt.xlabel("Decile")
plt.ylabel("Number of Students")
plt.show()
```

```python
## Misleading chart b/c bad y-axis
mentions = [500, 505]
years = [2013, 2014]

plt.bar([2013,2014] # x-values
        ,mentions # heights    
        ,0.8) # bar width
plt.title("Look at the \"Huge\" increase!")
plt.ylabel("Number of articles mentioning \"blockchain\"")
plt.xticks(years)

# if this is not present, matplotlib labels x-axis with 0, 1,
# and then add e in the corner (2.013e3)
plt.ticklabel_format(useOffset=False)

# misleading y-axis starts @ 490
plt.axis([2012.5,2014.5, 499,506])
plt.show()
```

### Line Charts

These are useful for trends

```python
variance     = [1,2,4,8,16,32,64,128,256]
bias_squared = [256,128,64,32,16,8,4,2,1]

# combine the 2 lists, column-wise
list(zip(variance,bias_squared))
```

```python
# calculate total_error
total_error = [x+y for x,y in zip(variance,bias_squared)]

# see index and values from 'variance' list
list(enumerate(variance))
```

```python
# get the x values for the plot via our indexes in 'variance'
xs = [i for i,_ in enumerate(variance)]

# make multiple calls .plot() show multiple series on same plot
plt.plot(xs, variance, "g-", label="variance") # green dashed
plt.plot(xs, bias_squared, "r-.", label="bias sq.") # red dot-dash
plt.plot(xs, total_error, "b:", label="total error") # blue dotted

# get a legend via our series' labels
plt.legend(loc=9) # put in top-center
plt.xlabel("Model Complexity")
plt.title("The Bias-Variance Trade-off")
plt.show()
```

*Side-note*: Notice how our total error is lowest when we find the pairing with a low bias *and* a low variance.

### Scatterplots

These are good for looking at relationships between paired sets of data.

```python
friends = [ 70, 65, 72, 63, 71, 64, 60, 64, 67]
minutes = [175, 170, 205, 120, 220, 130, 105, 145, 190]
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i']

# minutes vs. friends
plt.scatter(friends,minutes)

## label the points
# combine series as cols, go through each row
for label,friend_cnts,min_cnt in zip(labels,friends,minutes):
    # for each point, write the label
    plt.annotate(label, 
                 # put the label with each xy point (friend,minute)
                 xy=(friend_cnts,min_cnt), 
                 # slightly offset the point labels to be readable
                 xytext=(5,-5),
                 textcoords="offset points")
plt.title("Daily Minutes vs. Number of Friends")
plt.xlabel("Number of Friends")
plt.ylabel("Daily Minutes Spend on Site")
plt.show()
```

```python
# matplot lib can choose bad axes (non-comparable)
test_1_grades = [ 99, 90, 85, 97, 80]
test_2_grades = [100, 85, 60, 90, 70]

plt.scatter(test_1_grades, test_2_grades)
plt.title("Axes Aren't Comparable")
plt.xlabel("test 1 grade")
plt.ylabel("test 2 grade")
plt.show()
```

```python
# utilize call to plt.axis("equal") to more accurately display data
# matplot lib can choose bad axes (non-comparable)
test_1_grades = [ 99, 90, 85, 97, 80]
test_2_grades = [100, 85, 60, 90, 70]

plt.scatter(test_1_grades, test_2_grades)
plt.title("Axes Are Comparable")
plt.xlabel("test 1 grade")
plt.ylabel("test 2 grade")
plt.axis("equal")
plt.show()
```
