```python
with open('beer_origins.csv', 'rb') as f:
    result = chardet.detect(f.read())

beer_origin = pd.read_csv('./beer_origins.csv', encoding=result['encoding'])
beer_origin.head()
```

```python
all_beers = pd.merge(beer, beer_origin, on='beer_style')
all_beers.head()
```

```python
all_beers.tail()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

x = np.arange(len(all_beers['beer_origin'].unique()))
x1 = all_beers['beer_origin'].unique()
plt.bar(left = x, height = all_beers['beer_origin'].value_counts())
plt.xticks(x1)
```
